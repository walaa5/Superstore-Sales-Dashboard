```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import re
import numpy as np
import plotly.express as px
```

#### web scrapping Iso_alpha

```python
service = ChromeService(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.get('https://www.iban.com/country-codes')  
```

```python
country=driver.find_elements(By.TAG_NAME,'tr')
country[1].text
```

##### Process The data and Create DataFarme

```python
#get the total name
name1=re.sub('[0-9A-Z][0-9A-Z]+|\([a-z]*\)','',country[26].text)
print(name1)
```

```python
# get the name of country and the respected iso_alpha
country_name=[]
country_iso=[]
for c in country:
    data=(c.text).split()
    data_name=re.sub('[0-9A-Z][0-9A-Z]+|\([a-z]*\)','',c.text).strip()
    country_name.append(data_name)
    country_iso.append(data[-2])
```

```python
print(country_name[1:10])
print("length of country_Name: ",len(country_name))
print()
print(country_iso[1:10])
print("the length of country_iso: ",len(country_iso))
```

```python
#close the driver of webscrapping
driver.close()
```

```python
# create the dataFrame
data_list=list(zip(country_name[1:],country_iso[1:]))
data_iso=pd.DataFrame(data_list,columns=["country_name","iso_alpha"])
data_iso.head()
data_iso.to_csv("data_iso_webscrabbing.csv",header=True,index=False)
```

```python
flag=0
country_store_list=[]
count_world_list=[]
iso_country_store=[]
for country_store in country_list:
    for counntry_world in df.country_name:
        if (counntry_world.lower()).startswith(country_store.lower()):
            flag +=1
            country_store_list.append(country_store)
            count_world_list.append(counntry_world)
            print()
            print(country_store,"=======",counntry_world,"====",list(df.iso_alpha[df['country_name']==counntry_world])[0])
            print()
            iso_country_store.append(list(df.iso_alpha[df['country_name']==counntry_world])[0])        
```

#### get the orders_counts and profit in each market

```python
market_data=data.groupby('Market').agg({'Order ID':'count', 'Profit': 'sum'})
market_data.Profit=np.round(market_data.Profit)
market_data.head()
```

```python
markets=markets.join(market_data,how='left',on='Market')
markets
```

```python
markets.rename(columns={"Order ID":"Market_Order_Count","Profit":"Market_Profit"},inplace=True)
markets.head()
```

```python
markets.to_csv("markets.csv",index=False,header=True)
```
